```python
dummy_y = to_categorical(encoded_Y)
dummy_y
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(4, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1)
```

```python
import matplotlib.pyplot as plt

# График потерь
plt.figure()
plt.plot(history.history['loss'], label='Обучающая ошибка')
plt.plot(history.history['val_loss'], label='Валидационная ошибка')
plt.xlabel('Эпоха')
plt.ylabel('Ошибка')
plt.legend()
plt.show()

# График точности
plt.figure()
plt.plot(history.history['accuracy'], label='Обучающая точность')
plt.plot(history.history['val_accuracy'], label='Валидационная точность')
plt.xlabel('Эпоха')
plt.ylabel('Точность')
plt.legend()
plt.show()
```

```python
model_2hidden = Sequential()
model_2hidden.add(Dense(8, input_dim=4, activation='relu'))
model_2hidden.add(Dense(4, activation='relu'))
model_2hidden.add(Dense(3, activation='softmax'))
model_2hidden.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
